```python
# step1 : import library
import pandas as pd
```

```python
# step 2 : Fetch data
diabetes = pd.read_csv('Diabetes.csv')
diabetes.head()
```

```python
diabetes.columns
```

```python
# step 3 : find target and features
#define target(X)
Y=diabetes['diabetes']


# define features (Y)
# X = diabetes.drop(['diabetes'], axis=1)


X=diabetes[['pregnancies','glucose','diastolic','triceps','insulin','bmi','dpf','age']]
```

```python
# step 4 : Train test split
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.7,random_state=2529)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
# step 5 : Model Selection

# Logistic Model

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train,Y_train)  # Diabetes = 1 / ( 1 + e^(LinearRegression formula))  Formula
```

```python
# KNN model

# from sklearn.neighbors import KNeighborsClassifier

# model = KNeighborsClassifier()

# model.fit(X_train,Y_train)  # Diabetes = \|((x2-x1)^2 + (y2-y1)^2)  Formula
```

```python
# step 7 : Prediction
Y_pred = model.predict(X_test)
```

```python
# step 8 : Accuracy

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test,Y_pred)
accuracy = accuracy*100
error = 100 - accuracy
```

```python
print("Error : " , error)
```

```python
print("Accuracy : " , accuracy)
```
